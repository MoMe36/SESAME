```python
import numpy as np 
import scipy.signal as signal 
import scipy.optimize as optim
import matplotlib.pyplot as plt 
```

```python
g = signal.lti([1], [15,8,1])
plt.plot(*g.step())
```

```python
m = 10  # control horizon 
p = 20 # prediction horizon 
delta_t = 1 #sampling rate 
```

```python
t_continuous = np.linspace(0, p * delta_t, 1000) # closely spaced time points 
t_pred = np.arange(0, p*delta_t, delta_t) # discrete points for prediction horizon 
```

```python
tau_c = 1. 
r = 1. - np.exp(-t_pred/tau_c)
plt.plot(r)
```

```python
u  =np.ones(m)
x0 = np.zeros(g.to_ss().A.shape[0])
```

```python
def extend(u): 
    return np.concatenate([u, np.repeat(u[-1], p - m)])
```

```python
def predictions(u, t = t_pred, x0= x0): 
    t,y,x = signal.lsim(g, u, t, X0 = x0, interp = False)
    return y 
```

```python
plt.plot(t_pred ,predictions(extend(u)))
```

```python
def objective_func(u, x0 = x0): 
    y  =predictions(extend(u)) 
    return np.sum((r - y)**2) 
```

```python
objective_func(u)
```

```python
result = optim.minimize(objective_func, u)
u_opt = result.x 
print(result.fun)
```

```python
u_continuous = extend(u_opt)[((t_continuous - 0.01)//delta_t).astype(int)]
```

```python
def plot_output(u_continous, u_opt):
    plt.figure()
    plt.plot(t_continuous, u_continous)
    plt.xlim([0, delta_t*(p+1)])
    plt.figure()
    plt.plot(t_continuous, predictions(u_continous, t_continuous), label='Continuous response')
    plt.plot(t_pred, predictions(extend(u_opt)), '-o', label='Optimized response')
    plt.plot(t_pred, r, label='Set point')
    plt.legend()

plot_output(u_continuous, u_opt)
```
