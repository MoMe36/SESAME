```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
```

### MPC with GA 
* model: x_{t+1} = x_t + sell ? p_i * q_i : 0 
* cost: maximize returns 
* optim: genetic algorithm over sequence of actions
* Action: sequence of p_storage. Automatic sell action when storage > 80% 

First, create prod data and price data 

```python
N = 200 
prod_magn = 0.1
prod_freq = 6.7
prod_noise_magn = 0.01
price_magn = 40
price_freq = 0.7
price_offset = 2
price_noise_magn = 2 
x = np.linspace(-np.pi, np.pi, N)

production = ((np.sin(x * prod_freq) + 1)*0.5) * prod_magn
production += np.random.normal(0,prod_noise_magn, size = production.shape)

price = ((np.sin(x * price_freq) + price_offset)*0.5) * price_magn
price += np.random.normal(0,price_noise_magn, size = price.shape)

```

```python
data = pd.DataFrame(np.hstack([production.reshape(-1,1), price.reshape(-1,1)]), columns = 'prod,price'.split(','))
data.head()
```

```python
f, axes = plt.subplots(2,1, figsize = (10,10))
axes[0].plot(data['prod'])
axes[0].set_title('Production', weight = 'bold')
axes[1].plot(data['price'])
axes[1].set_title('Price', weight = 'bold')
```

```python
def sim_model(soc, current_prod, p_storage_i, max_soc, sale_th = 0.8, efficiency = 0.7): 
    next_soc = soc + current_prod * p_storage_i * efficiency
    
    qty_to_sell  = 0. 
    if next_soc / max_soc > sale_th: 
        qty_to_sell += next_soc - max_soc * 0.1 # Battery returns to 10% of max load
        next_soc -= qty_to_sell 
        
    qty_to_sell += current_prod * (1. - p_storage_i)
    return next_soc, qty_to_sell 

def compute_fitness(action_sequence, data, idx, horizon_length):
        return 0 
```

```python
def get_action(idx, **kwargs): 
    if kwargs['strategy'] == 'random': 
        return np.random.uniform(0.,1.)
    else: 
        return 0. 
```

```python
def show_results(records): 
    f, axes = plt.subplots(2,1, figsize = (10,8))
    axes = axes.flatten()
    axes[0].plot(records.SOC)
    axes[0].set_title('SoC', weight = 'bold')
    axes[1].plot(records.Equity, label = 'Individual')
    axes[1].plot(records.Equity.cumsum(), label = 'Cumulative')
    axes[0].set_title('Economic returns', weight = 'bold')
```

```python
horizon = 15
sim_params = {'strategy': 'random'}
soc = 0. 
soc_max = 2. #MW (?)
equity = 0. 

record_cols = ['SOC', 'Equity'] 
records = pd.DataFrame(data = [[soc/soc_max, equity]], columns = record_cols)
for i in range(data.shape[0]):
    p_storage = get_action(i, **sim_params)
    soc, qty_to_sell = sim_model(soc, data['prod'][i], p_storage, soc_max)
    sale_return = qty_to_sell * data.price[i]
    records = records.append(pd.DataFrame([[soc, sale_return]], columns = record_cols))
records.reset_index(drop = True, inplace = True)
```

```python
show_results(records)
```
